```python
from sklearn.datasets import fetch_california_housing # imports the function to load the California housing dataset

housing = fetch_california_housing() # loads the California housing dataset into the variable housing
```

```python
# imports pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# defines the url for the Boston housing dataset
data_url = "http://lib.stat.cmu.edu/datasets/boston"
# reads the dataset from the url into a pandas DataFrame
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
# extracts the data and target variables from the DataFrame
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# prints the data and target variables
print(data)
print(target)
```

```python
#Organising and printing first 5
boston_df=pd.DataFrame(housing.data,columns=housing.feature_names)
boston_df.head()
```
