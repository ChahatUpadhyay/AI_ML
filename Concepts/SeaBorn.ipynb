Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# total bill vs tip dataset
# Check Seaborn documentation for builtin functions
tips=sns.load_dataset('tips')
tips.head()
```

```python
# seeting a theme for the plots
sns.set_theme()# only run it ones
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',col='time', hue='sex', style='smoker',size='size')
```

```python
# Iris Dataset importing datasets
# helps in predicting about the species of the Iris
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

```python
#using petal width
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species')
```

```python
#Import the required libraries
from sklearn.datasets import fetch_california_housing # Use fetch_california_housing instead of load_boston

# Load the California housing dataset
housing = fetch_california_housing()

# Convert the dataset to a Pandas DataFrame
housing_df = pd.DataFrame(data=housing.data, columns=housing.feature_names)

# Add the target variable to the DataFrame
housing_df['target'] = housing.target

# Display the first few rows of the DataFrame
housing_df.head()
```

Distribute part

```python
sns.displot(data=housing_df, x='MedInc', kde=True)
```

```python
sns.displot(housing_df['HouseAge'])
```

```python
#Corerelation +ve and -ve
# Heat map
plt.figure(figsize=(10,10))
sns.heatmap(housing_df.corr(),annot=True)
# gives r for all column to all all column
```
