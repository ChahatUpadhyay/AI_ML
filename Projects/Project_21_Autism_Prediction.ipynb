```python
# data processing
# lets first hide the outliers

# function to replace the outliers with median
def replace_outliers_with_median(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  median = df[column].median()

  # replace outliers with median value
  df[column] = df[column].apply(lambda x: median if x < lower_bound or x > upper_bound else x)

  return df
```

```python
# replace outliers in the "age" column
df = replace_outliers_with_median(df, "age")

# replace outliers in the "result" column
df = replace_outliers_with_median(df, "result")
```

```python
# SMOTE(Synthetic Minority Oversampling) from imblearn
smote=SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
y_train_resampled.value_counts()
```

**5. Model Training**

```python
# dictionary of classifiers
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}
```

```python

# dictionary to store the cross validation results
cv_scores = {}

# perform 5-fold cross validation for each model
for model_name, model in models.items():
  print(f"Training {model_name} with default parameters...")
  scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=5, scoring="accuracy")
  cv_scores[model_name] = scores
  print(f"{model_name} Cross-Validation Accuracy: {np.mean(scores):.2f}")
  print("-"*50)
```

```python
cv_scores
```

**6. Model Selection & Hyperparameter Tuning**

```python
# Initializing models
decision_tree = DecisionTreeClassifier(random_state=42)
random_forest = RandomForestClassifier(random_state=42)
xgboost_classifier = XGBClassifier(random_state=42)
```

```python
# Hyperparameter grids for RandomizedSearchCV

param_grid_dt = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 10, 20, 30, 50, 70],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}


param_grid_rf = {
    "n_estimators": [50, 100, 200, 500],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "bootstrap": [True, False]
}


param_grid_xgb = {
    "n_estimators": [50, 100, 200, 500],
    "max_depth": [3, 5, 7, 10],
    "learning_rate": [0.01, 0.1, 0.2, 0.3],
    "subsample": [0.5, 0.7, 1.0],
    "colsample_bytree": [0.5, 0.7, 1.0]
}
```

```python
# hyperparameter tunig for 3 tree based models

# the below steps can be automated by using a for loop or by using a pipeline

# perform RandomizedSearchCV for each model
random_search_dt = RandomizedSearchCV(estimator=decision_tree, param_distributions=param_grid_dt, n_iter=20, cv=5, scoring="accuracy", random_state=42)
random_search_rf = RandomizedSearchCV(estimator=random_forest, param_distributions=param_grid_rf, n_iter=20, cv=5, scoring="accuracy", random_state=42)
random_search_xgb = RandomizedSearchCV(estimator=xgboost_classifier, param_distributions=param_grid_xgb, n_iter=20, cv=5, scoring="accuracy", random_state=42)
```

```python
# fit the models
random_search_dt.fit(X_train_resampled, y_train_resampled)
random_search_rf.fit(X_train_resampled, y_train_resampled)
random_search_xgb.fit(X_train_resampled, y_train_resampled)
```

```python
# Get the model with best score

best_model = None
best_score = 0

if random_search_dt.best_score_ > best_score:
  best_model = random_search_dt.best_estimator_
  best_score = random_search_dt.best_score_

if random_search_rf.best_score_ > best_score:
  best_model = random_search_rf.best_estimator_
  best_score = random_search_rf.best_score_

if random_search_xgb.best_score_ > best_score:
  best_model = random_search_xgb.best_estimator_
  best_score = random_search_xgb.best_score_

```

```python
print(f"Best Model: {best_model}")
print(f"Best Cross-Validation Accuracy: {best_score:.2f}")
```

```python
# save the best model
with open("Autism_best_model.pkl", "wb") as f:
  pickle.dump(best_model, f)

# from joblib import dump
# # Save the model
# dump(best_model, "Autism_best_model.joblib")
```

```python
# evaluate on test data
y_test_pred = best_model.predict(X_test)
print("Accuracy score:\n", accuracy_score(y_test, y_test_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
```

```python
import pickle
# Load the saved model
with open("Autism_best_model.pkl", "rb") as f:
    loaded_model = pickle.load(f)
# Load the saved encoders
with open("encoders.pkl", "rb") as f:
    loaded_encoders = pickle.load(f)


def predict_autism(input_data):
    # Create a DataFrame from the input data
    input_df = pd.DataFrame([input_data])

    # Get the training data's feature names
    training_features = loaded_model.feature_names_in_  # Assuming loaded_model has feature_names_in_ attribute

    # Reorder or add missing columns in input_df to match training features
    input_df = input_df.reindex(columns=training_features, fill_value=0)  # Fill missing columns with 0

    # Apply label encoding using the loaded encoders
    for column, encoder in loaded_encoders.items():
        if column in input_df.columns:
            input_df[column] = encoder.transform(input_df[column])

    # Make prediction using the loaded model
    prediction = loaded_model.predict(input_df)
    return prediction[0]  # Return the prediction (0 or 1)


# Example usage:
# Assuming you have a new data point in a dictionary format
# new_data_point = {
#     'A1_Score': 1,
#     'A2_Score': 1,
#     'A3_Score': 1,
#     'A4_Score': 0,
#     'A5_Score': 1,
#     'A6_Score': 1,
#     'A7_Score': 1,
#     'A8_Score': 1,
#     'A9_Score': 1,
#     'A10_Score': 1,
#     'age': 18,  # Assuming '18 and more' means 18+
#     'gender': 'm',
#     'ethnicity': 'White-European',
#     'jaundice': 'no',
#     'austim': 'yes',
#     'contry_of_res': 'United States',
#     'used_app_before': 'no',
#     'relation': 'Self',
#     'result': 0
# }


new_data_point = {
    'A1_Score': 1,
    'A2_Score': 1,
    'A3_Score': 1,
    'A4_Score': 1,
    'A5_Score': 0,
    'A6_Score': 0,
    'A7_Score': 1,
    'A8_Score': 1,
    'A9_Score': 0,
    'A10_Score': 0,
    'age': 5,
    'gender': 'f',
    'ethnicity': 'White-European',
    'jaundice': 'no',
    'austim': 'no',
    'contry_of_res': 'United States',
    'used_app_before': 'no',
    'relation': 'Self',
    'result': 5
}

# Get the prediction
prediction = predict_autism(new_data_point)
print(prediction)
if prediction == 0:
  print("The person has no Autism")
else:
  print("The person has Autism")
```
