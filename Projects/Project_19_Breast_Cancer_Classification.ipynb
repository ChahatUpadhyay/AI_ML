```python
# importing lobraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#dataframe
df=pd.read_csv('/content/data_.csv')
```

```python
df.head()
```

```python
# Adding label column for each value consider replacing B with 1 and M with 0 on diagnosis column
df['label'] = df['diagnosis'].map({'B': 1, 'M': 0})
```

```python
#checking for null values
df.isnull().sum()
```

```python
df.head()
```

```python
#delete unmaned:32 column
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
# checking the distribution of Target Varibale
df['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
df.head()
```

```python
# Calculate the mean only for numeric columns
numeric_df = df.select_dtypes(include=np.number)
grouped_data = numeric_df.groupby('label').mean()
print(grouped_data)
```

```python
X = df.drop(columns=['label', 'diagnosis'], axis=1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# model traninig
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
```

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```
